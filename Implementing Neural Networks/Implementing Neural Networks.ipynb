# Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,Normalizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
import tensorflow as tf
```

### Data Loading and observing

```python
# loading dataset
dataset = pd.read_csv('life_expectancy.csv')
# observing first five entries
dataset.head()
```

```python
# datatypes of columns
print(dataset.dtypes)
```

```python
# Summary stat of dataset
dataset.describe()
```

```python
dataset=dataset.drop(['Country'],axis=1)
```

```python
dataset.head()
```

```python
labels = dataset.iloc[:,-1]
```

```python
labels.head()
```

```python
features = dataset.iloc[:, 0:-1]
features
```

## Data Preprocessing

```python
# one hot encoding for categorical columns
features = pd.get_dummies(features)
features.head()
```

## Split data into training and test sets

```python
features_train,features_test,labels_train,labels_test = train_test_split(features,labels, test_size=0.2,random_state=42)
```

### Normalizing Numerical Features

```python
numerical_features = features.select_dtypes(include=['float64','int64'])
numerical_columns = numerical_features.columns

ct = ColumnTransformer([("standardize", StandardScaler(), numerical_columns)], remainder='passthrough')
```

```python
features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.transform(features_test)
```

## Building the model

```python
my_model = Sequential()
```

```python
input = tf.keras.layers.InputLayer(input_shape=(dataset.shape[1],))
my_model.add(input)
```

```python
hidden = tf.keras.layers.Dense(8, activation='relu')
my_model.add(hidden)
```

```python
output = tf.keras.layers.Dense(1)
my_model.add(output)
```

```python
# summary of the model
my_model.summary()
```

##### Optimizer

```python
opt = Adam(learning_rate = 0.01)
```

## Compile Model

```python
my_model.compile(loss ='mse', metrics=['mae'],optimizer = opt)
```

##  Fit and Evaluate the Model

```python
my_model.fit(features_train,labels_train,epochs = 40, batch_size=1,verbose = 1)
```

## Evaluating model

```python
res_mse,res_mae = my_model.evaluate(features_test,labels_test,verbose = 0)
```
